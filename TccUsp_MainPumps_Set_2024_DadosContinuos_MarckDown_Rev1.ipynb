(2023) avalia uma s√©rie temporal como sendo um conjunto de vari√°veis ordenadas em fun√ß√£o do tempo, e que o estudo e modelagem destes dados, √© fundamental para compreender o comportamento e realizar previs√µes sobre determinados dados. Ele orienta para a necessidade de realiza√ß√£o de um estudo do tipo de serie temporal que est√° sendo objeto de an√°lise e proje√ß√£o, e que em caso de observar que os dados possuem tend√™ncia estoc√°stica, √© preciso avaliar pois isso pode tornar as previs√µes imprecisas conforme se afastam do √∫ltimo ponto da amostra. Para tanto, ele orienta a utiliza√ß√£o do procedimento para a avalia√ß√£o da estacionariedade da s√©rie, por meio do teste de Dickey-Fuller, como sendo um dos testes de raiz unit√°ria mais tradicionais. A seguir, temos na Eq (1, 2 e 3) representando o teste de Dickey-Fuller.

y_t=‚àÖy_(t-1)+u_t                                                                                                                                   (1)

y_t=Œ≤_1+‚àÖy_(t-1)+ u_t                                                                                                                              (2)

y_t=Œ≤_1+Œ≤_2 t+ ‚àÖy_(t-1)+ u_t                                                                                                                       (3)


em que ùúô √© um par√¢metro a ser estimado, ùë¢ùë° √© um processo de ru√≠do branco, que se caracteriza como uma sequ√™ncia de vari√°veis aleat√≥rias independentes e identicamente distribu√≠das (iid), com m√©dia zero e vari√¢ncia constante (ùë¢ùë°‚àºùëÖùêµ (0, ùúé2)), ùõΩ1√© uma constante que representa o intercepto e ùõΩ2 √© um efeito de tend√™ncia. O teste considera como hip√≥tese nula, ùêª0 ‚à∂ ùúô= 1, a presen√ßa de raiz unit√°ria e como hip√≥tese alternativa, ùêª1‚à∂ùúô <|1|, a s√©rie sendo estacion√°ria.

Silveira et al. (2022) Na presen√ßa de estacionariedade, a fun√ß√£o amostral que descreve o processo gerador dos da dos tem a mesma forma em todos os instantes, facilitando a identifica√ß√£o de estimativas dos par√¢metros desconhecidos dos modelos especificados, e relaciona alguns avan√ßos nos testes de ADF e sugere a utiliza√ß√£o do c√°lculo do valor- p como determinado a partir da estat√≠stica œÑ no teste ADF, e poss√≠veis interfer√™ncias no resultado.

Em nosso estudo, definiremos o valor-p (p-value) e em caso de valores pr√≥ximo de zero, indica que h√° forte evid√™ncia estat√≠stica contra a hip√≥tese nula de que a s√©rie temporal √© n√£o estacion√°ria, ou seja, √© poss√≠vel concluir com confian√ßa que a s√©rie temporal √© estacion√°ria.

Aileen Nielsen (2021) explica, que se a s√©rie temporal √© estacion√°ria simplifica significativamente a an√°lise e modelagem pois indica que suas propriedades estat√≠sticas, como m√©dia, vari√¢ncia e autocorrela√ß√£o, permanecem constantes ao longo do tempo. Isso permite aplicar uma variedade de t√©cnicas de modelagem e previs√£o com mais confian√ßa.

Aliene Nielsen (2021) separa os modelos estat√≠sticos de series temporais entre, est√°ticos, modelos de esparo de estado e aprendizado de m√°quina, sendo que em nosso estudo, utilizaremos alguns destes descritos a seguir, sendo que a avalia√ß√£o e defini√ß√£o do modelo mais aderente estar√° condicionada aos resultados de MAE (), MSE e R2 para os seguintes modelos, conforme descrito por Aur√©lien G√©ron (2021) como sendo alguns dos m√©todos estat√≠sticos desenvolvidos para series temporais:

Regress√£o Linear Simples ou M√∫ltipla: Aur√©lien G√©ron (2021) A regress√£o linear √© um modelo simples que pode ser eficaz se houver uma rela√ß√£o linear direta entre as vari√°veis de entrada e sa√≠da.

Modelos Autorregressivos (AR): Modelos autorregressivos consideram a rela√ß√£o entre uma observa√ß√£o atual e observa√ß√µes passadas. Eles s√£o √∫teis quando h√° depend√™ncia temporal nas s√©ries temporais.

M√©dias M√≥veis (MA): Modelos de m√©dias m√≥veis consideram a rela√ß√£o entre uma observa√ß√£o e um erro residual das observa√ß√µes passadas. Eles podem ser combinados com modelos AR para formar modelos ARMA.

Modelos ARIMA (Autoregressive Integrated Moving Average): Aur√©lien G√©ron (2021) avalia que os modelos ARIMA combinam componentes de regress√£o autorregressiva, m√©dias m√≥veis e diferencia√ß√£o para modelar s√©ries temporais estacion√°rias ou com tend√™ncias conhecidas.

Suaviza√ß√£o Exponencial (Exponential Smoothing): Modelos recomendados para suavizar s√©ries temporais e capturar padr√µes sazonais.

Redes Neurais Recorrentes (RNNs): Como dito por Aur√©lien G√©ron (2021), modelos de redes neurais recorrentes, como LSTM e GRU, s√£o eficazes para capturar depend√™ncias de longo prazo em s√©ries temporais, sendo a LSTM tem resultados melhores, contudo a GRU e mais r√°pida no treinamento mesmo que em determinados estudos o resultado da GRU tenha sido melhor.

State Space Models: Modelos recomendados para modelar s√©ries temporais que podem ser decompostas em estados latentes.

Tamb√©m como conclu√≠do por Aliene Nielsen (2021), modelos complicados nem sempre s√£o os melhores, pois o custo-benef√≠cio compensa o uso de recursos computacionais adicionais requeridos para sua opera√ß√£o.

Em uma an√°lise e defini√ß√£o de um modelo, e importante pensar se o tempo de treinamento adicional para operar com um modelo de aprendizagem de m√°quinas complexo valem a pena.

Aliene Nielsen (2021) tamb√©m apresenta as diferen√ßas entre os modelos de series temporais univariadas e multivariadas, ou seja, series temporais univariadas possuem apenas uma vari√°vel medida ao longo do tempo, j√° as series temporais multivariadas s√£o series com m√∫ltiplas vari√°veis medidas a cada timestamp.
```python
from zipfile import ZipFile
from io import BytesIO
import requests
import os
import pandas as pd

# URL do reposit√≥rio no GitHub
repo_url = 'https://github.com/CidClayQuirino/rnn-component-lIfe-cycle/archive/main.zip'
dataframes = []

# Baixe e extraia o arquivo zip do reposit√≥rio
response = requests.get(repo_url)
with ZipFile(BytesIO(response.content)) as zip_file:
    zip_file.extractall()

# Diret√≥rio onde os arquivos .xlsx foram extra√≠dos
extracted_dir = 'rnn-component-lIfe-cycle-main'

# Loop pelos arquivos no diret√≥rio extra√≠do
for arquivo in os.listdir(extracted_dir):
    if arquivo.endswith('.xlsx'):
        # Construa o caminho completo para o arquivo
        caminho_completo = os.path.join(extracted_dir, arquivo)

        # Leia o arquivo Excel e adicione-o √† lista de DataFrames
        df = pd.read_excel(caminho_completo)

        # Adicione uma coluna 'TagComp' contendo o nome do arquivo sem a extens√£o
        df['nome_arquivo'] = os.path.splitext(arquivo)[0]

        # Adicione o DataFrame √† lista
        dataframes.append(df)

# Concatene todos os DataFrames em um √∫nico DataFrame
BDadosTemp = pd.concat(dataframes, ignore_index=True)

# Filtrar linhas onde todas as colunas n√£o t√™m valor zero
BDadosTemp = BDadosTemp[(BDadosTemp != 0).all(axis=1)]

# Use .loc[] para evitar o SettingWithCopyWarning
BDadosTemp.loc[:, 'Value'] = BDadosTemp['Value'].round(1)
BDadosTemp = BDadosTemp.rename(columns={'Tag': 'Parametro', 'nome_arquivo': 'NmeComp'})
```

# Teste de Shapiro-Wilk (n < 30)

# Teste de Shapiro-Francia (n >= 30)
# Carregamento da fun√ß√£o 'shapiro_francia' do pacote 'statstests.tests'

```python
import pandas as pd
from scipy.stats import shapiro

# Supondo que df_MainPumpsTemp seja o DataFrame com os dados de temperatura para ambas as bombas P1 e P2

# Filtrar os dados para MainPumpP1 e MainPumpP2
data_p1 = df_MainPumpsTemp[df_MainPumpsTemp['NmeComp'] == 'MainPumpP1']['Value']
data_p2 = df_MainPumpsTemp[df_MainPumpsTemp['NmeComp'] == 'MainPumpP2']['Value']

# Teste de Shapiro-Wilk para MainPumpP1
stat_p1, p_value_p1 = shapiro(data_p1)
print(f'Shapiro-Wilk Test for MainPumpP1: W={stat_p1:.5f}, p-value={p_value_p1:.6f}')
alpha = 0.05
if p_value_p1 > alpha:
    print('N√£o se rejeita H0 para MainPumpP1 - Distribui√ß√£o aderente √† normalidade')
else:
    print('Rejeita-se H0 para MainPumpP1 - Distribui√ß√£o n√£o aderente √† normalidade')

# Teste de Shapiro-Wilk para MainPumpP2
stat_p2, p_value_p2 = shapiro(data_p2)
print(f'Shapiro-Wilk Test for MainPumpP2: W={stat_p2:.5f}, p-value={p_value_p2:.6f}')
if p_value_p2 > alpha:
    print('N√£o se rejeita H0 para MainPumpP2 - Distribui√ß√£o aderente √† normalidade')
else:
    print('Rejeita-se H0 para MainPumpP2 - Distribui√ß√£o n√£o aderente √† normalidade')
```

A seguir as Figuras 23a e b, mostra o teste de Dickey-Fuller, utilizando a biblioteca do python ‚Äúimport adfuller‚Äù para identificar o padr√£o estacion√°rio ou n√£o dos dados. Os resultados dessa an√°lise para a amostra de dados do estudo, apresentaram um Valor-p = 0.000003 para a Bomba P1 e de Valor-p = 0.000009 para a Bomba P2, est√£o muito pr√≥ximo de zero, ou seja, indicando que h√° forte evid√™ncia estat√≠stica para rejeitar a hip√≥tese nula de que a s√©rie √© n√£o estacion√°ria, ou seja, este resultado tanto para a Base de dados da Bomba 1 quanto para a Bomba 2, sugere-se que a s√©rie temporal seja estacion√°ria.

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# Separa√ß√£o dos dados de temperatura das Bombas P1 e P2
Df_Pump1 = df_MainPumpsTemp[df_MainPumpsTemp['NmeComp'] == 'MainPumpP1']
# Primeiro, vamos filtrar a s√©rie temporal desejada (por exemplo, usando a coluna 'Parametro' para selecionar)
serie_temporalP1 = Df_Pump1['Value']
# Fun√ß√£o para testar estacionariedade da s√©rie temporal
def test_stationarity(timeseries):
    # Verificar se a s√©rie temporal n√£o est√° vazia
    if timeseries.empty:
        print("S√©rie temporal est√° vazia.")
        return
    # Estat√≠sticas de rolamento
    rolmean = timeseries.rolling(window=12).mean()
    rolstd = timeseries.rolling(window=12).std()
    # Plotar estat√≠sticas de rolamento
    import matplotlib.pyplot as plt
    plt.plot(timeseries, color='blue',label='Original')
    plt.plot(rolmean, color='red', label='M√©dia M√≥vel')
    plt.plot(rolstd, color='black', label = 'Desvio Padr√£o M√≥vel')
    plt.legend(loc='best')
    plt.title('Estat√≠sticas de Rolamento')
    plt.show()
    # Teste de Dickey-Fuller:
    print('Resultados do Teste Dickey-Fuller:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Estat√≠stica do Teste','Valor-p','#Lags Usados','N√∫mero de Observa√ß√µes Usadas'])
    for key,value in dftest[4].items():
        dfoutput['Valor Cr√≠tico (%s)'%key] = value
    print(dfoutput)
# Aplicar teste de estacionariedade √† s√©rie temporal selecionada
test_stationarity(serie_temporalP1)
```

Aplica√ß√£o: Um modelo AR pode ser utilizado se os valores passados do pr√≥prio tempo explicam bem os valores futuros. O MA √© √∫til se os erros passados t√™m uma influ√™ncia significativa.
3. Modelos de Suaviza√ß√£o Exponencial (Exponential Smoothing, Holt-Winters):
Justificativa: Embora os res√≠duos n√£o sigam uma distribui√ß√£o normal, os modelos de suaviza√ß√£o exponencial, como Holt-Winters, podem ser bastante eficazes para s√©ries temporais onde a m√©dia e a tend√™ncia variam suavemente ao longo do tempo.
Aplica√ß√£o: Se houver uma tend√™ncia ou sazonalidade clara, modelos de suaviza√ß√£o exponencial podem ser aplicados para capturar essas din√¢micas.
4. Modelos de Redes Neurais Recorrentes (RNNs, LSTM, GRU):
Justificativa: Se os res√≠duos n√£o seguem uma distribui√ß√£o normal, os modelos baseados em redes neurais, como LSTM (Long Short-Term Memory) ou GRU (Gated Recurrent Unit), podem ser eficazes. Esses modelos n√£o assumem normalidade e s√£o capazes de capturar padr√µes complexos e depend√™ncias de longo prazo em s√©ries temporais.
Aplica√ß√£o: LSTM e GRU s√£o recomendados para dados sequenciais complexos, especialmente se h√° evid√™ncia de que os modelos lineares tradicionais (como ARIMA) n√£o capturam completamente as din√¢micas da s√©rie.
5. Modelos de M√°quinas de Vetores de Suporte (SVR):
Justificativa: SVR (Support Vector Regression) pode ser utilizado para s√©ries temporais, especialmente quando os dados n√£o seguem uma distribui√ß√£o normal. SVR √© robusto a outliers e pode capturar rela√ß√µes n√£o lineares.
Aplica√ß√£o: SVR pode ser ajustado com t√©cnicas como RandomizedSearchCV para otimizar seus hiperpar√¢metros e encontrar um modelo que se ajuste bem aos dados.
Conclus√£o e Recomenda√ß√£o:
Dado que a s√©rie √© estacion√°ria e que os res√≠duos n√£o seguem uma distribui√ß√£o normal, modelos como ARIMA/SARIMA/SARIMAX s√£o uma boa escolha inicial. Se esses modelos n√£o capturarem adequadamente as din√¢micas da s√©rie, voc√™ pode considerar modelos de suaviza√ß√£o exponencial ou redes neurais recorrentes (LSTM, GRU), que s√£o robustos a desvios de normalidade e podem capturar padr√µes mais complexos.

Se houver evid√™ncias de n√£o linearidade, SVR com ajuste adequado de hiperpar√¢metros pode ser uma alternativa vi√°vel. Em todos os casos, √© importante validar o modelo usando dados de teste ou valida√ß√£o cruzada para garantir que ele generalize bem para novos dados.

A defini√ß√£o pela t√©cnica de AST&P (Advanced Statistical Techniques & Procedures) para an√°lise de dados e tomada de decis√£o foram feitos utilizando t√©cnicas estat√≠sticas cl√°ssicas e avan√ßadas. Dentre estes, o modelo mais aderente para trabalhar com s√©ries temporais lineares, depende de v√°rios fatores, incluindo a estrutura dos dados, a quantidade de dados dispon√≠veis, a presen√ßa de tend√™ncias ou sazonalidades, entre outros, contudo os modelos que foram testados para avaliar os resultados nessa s√©rie temporal de temperatura foram:

‚Ä¢	Regress√£o Linear Simples ou M√∫ltipla: A regress√£o linear √© um modelo simples que pode ser eficaz se houver uma rela√ß√£o linear direta entre as vari√°veis de entrada e sa√≠da.
‚Ä¢	Modelos Autoregressivos (AR): Modelos autorregressivos consideram a rela√ß√£o entre uma observa√ß√£o atual e observa√ß√µes passadas. Eles s√£o √∫teis quando h√° depend√™ncia temporal nas s√©ries temporais.
‚Ä¢	M√©dias M√≥veis (MA): Modelos de m√©dias m√≥veis consideram a rela√ß√£o entre uma observa√ß√£o e um erro residual das observa√ß√µes passadas. Eles podem ser combinados com modelos AR para formar modelos ARMA.
‚Ä¢	Modelos ARIMA (Autoregressive Integrated Moving Average): Modelos ARIMA combinam componentes de regress√£o autorregressiva, m√©dias m√≥veis e diferencia√ß√£o para modelar s√©ries temporais estacion√°rias ou com tend√™ncias conhecidas.
‚Ä¢	Suaviza√ß√£o Exponencial (Exponential Smoothing): Modelos recomendados para suavizar s√©ries temporais e capturar padr√µes sazonais.
‚Ä¢	Redes Neurais Recorrentes (RNNs): Modelos de redes neurais recorrentes, como LSTM e GRU, s√£o eficazes para capturar depend√™ncias de longo prazo em s√©ries temporais.
‚Ä¢	State Space Models: Modelos recomendados para modelar s√©ries temporais que podem ser decompostas em estados latentes.

```python
# Importar as bibliotecas necess√°rias
import pandas as pd
import matplotlib.pyplot as plt
from arch import arch_model

# Fun√ß√£o para aplicar o modelo GARCH e plotar os resultados
def aplicar_garch(df_pump, pump_name):
    # Configurar a coluna de 'Timestamp' como √≠ndice para trabalhar com s√©ries temporais
    df_pump.set_index('Timestamp', inplace=True)

    # Selecionar a coluna 'Value' para a an√°lise GARCH
    serie_temporal = df_pump['Value']

    # Plotar a s√©rie temporal para visualiza√ß√£o
    plt.figure(figsize=(10, 5))
    plt.plot(serie_temporal, label=f'S√©rie Temporal de Temperatura - {pump_name}')
    plt.title(f'S√©rie Temporal de Temperatura da {pump_name}')
    plt.xlabel('Timestamp')
    plt.ylabel('Value')
    plt.legend()
    plt.show()

    # Aplicar o modelo GARCH (GARCH(1,1))
    garch_model = arch_model(serie_temporal, vol='Garch', p=1, q=1)

    # Ajustar o modelo
    garch_fitted = garch_model.fit(disp='off')

    # Exibir o sum√°rio dos resultados do GARCH
    print(f'Sum√°rio do GARCH para {pump_name}:')
    print(garch_fitted.summary())

    # Fazer previs√µes com o modelo ajustado
    forecasts = garch_fitted.forecast(horizon=10)

    # Exibir a previs√£o da vari√¢ncia condicional
    print(f'Previs√£o da Vari√¢ncia Condicional para {pump_name}:')
    print(forecasts.variance[-1:])

    # Plotar a vari√¢ncia condicional prevista
    plt.figure(figsize=(10, 5))
    plt.plot(forecasts.variance.iloc[-1, :], label=f'Previs√£o da Vari√¢ncia Condicional - {pump_name}')
    plt.title(f'Previs√£o da Vari√¢ncia Condicional com GARCH(1,1) - {pump_name}')
    plt.xlabel('Per√≠odo')
    plt.ylabel('Vari√¢ncia')
    plt.legend()
    plt.show()

# Separa√ß√£o dos dados de temperatura das Bombas P1 e P2
Df_Pump1 = df_MainPumpsTemp[df_MainPumpsTemp['NmeComp'] == 'MainPumpP1'].sort_values(by='Timestamp')
Df_Pump2 = df_MainPumpsTemp[df_MainPumpsTemp['NmeComp'] == 'MainPumpP2'].sort_values(by='Timestamp')

# Aplicar o modelo GARCH para MainPumpP1
aplicar_garch(Df_Pump1, 'MainPumpP1')

# Aplicar o modelo GARCH para MainPumpP2
aplicar_garch(Df_Pump2, 'MainPumpP2')

```

```python
import numpy as np
import pandas as pd
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM, GRU
from keras.optimizers import Adam
from github import Github
import base64

# Vari√°veis do GitHub
seu_token = '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'
seu_usuario = 'CidClayQuirino'
seu_repositorio = 'rnn-component-lIfe-cycle'
file_path = 'SD_ResultGARCH.csv'

# Conectar ao GitHub usando o token de acesso
g = Github(seu_token)
repo = g.get_repo(f"{seu_usuario}/{seu_repositorio}")

# Fun√ß√£o para preparar dados em forma de sequ√™ncias para LSTM/GRU
def prepare_data(series, n_steps):
    X, y = [], []
    for i in range(len(series)):
        end_ix = i + n_steps
        if end_ix > len(series)-1:
            break
        seq_x, seq_y = series[i:end_ix], series[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)

# Fun√ß√£o para treinar modelos e salvar resultados
def treinar_modelos(df_pump, pump_name, n_steps=10):
    # Selecionar a s√©rie temporal 'Value' como vari√°vel de an√°lise
    serie_temporal = df_pump['Value'].values.reshape(-1, 1)
    y_true = serie_temporal

    # Escalar os dados para [0, 1]
    scaler = MinMaxScaler()
    serie_temporal_scaled = scaler.fit_transform(serie_temporal)

    # Criar sequ√™ncias de dados
    X, y = prepare_data(serie_temporal_scaled, n_steps)

    # Dividir os dados em treino e teste
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

    # 1. Treinamento e previs√£o com SVR
    X_train_svr = X_train.reshape((X_train.shape[0], -1))
    X_test_svr = X_test.reshape((X_test.shape[0], -1))

    svr_model = SVR(kernel='rbf', C=1.0, epsilon=0.1)
    svr_model.fit(X_train_svr, y_train)
    y_pred_svr = svr_model.predict(X_test_svr)
    y_pred_svr = scaler.inverse_transform(y_pred_svr.reshape(-1, 1))

    # 2. Treinamento e previs√£o com LSTM
    lstm_model = Sequential()
    lstm_model.add(LSTM(50, activation='relu', input_shape=(n_steps, 1)))
    lstm_model.add(Dense(1))
    lstm_model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
    lstm_model.fit(X_train, y_train, epochs=50, verbose=0)
    y_pred_lstm = lstm_model.predict(X_test, verbose=0)
    y_pred_lstm = scaler.inverse_transform(y_pred_lstm)

    # 3. Treinamento e previs√£o com GRU
    gru_model = Sequential()
    gru_model.add(GRU(50, activation='relu', input_shape=(n_steps, 1)))
    gru_model.add(Dense(1))
    gru_model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
    gru_model.fit(X_train, y_train, epochs=50, verbose=0)
    y_pred_gru = gru_model.predict(X_test, verbose=0)
    y_pred_gru = scaler.inverse_transform(y_pred_gru)

    # Definir os valores reais y_true para o conjunto de teste
    y_true_test = scaler.inverse_transform(y_test.reshape(-1, 1))

    # Criar um DataFrame para armazenar os resultados
    resultados = pd.DataFrame({
        'y_true': y_true_test.flatten(),
        f'y_pred_svr_{pump_name}': y_pred_svr.flatten(),
        f'y_pred_lstm_{pump_name}': y_pred_lstm.flatten(),
        f'y_pred_gru_{pump_name}': y_pred_gru.flatten()
    })

    return resultados

# Treinar os modelos para MainPumpP1
Df_Pump1 = df_MainPumpsTemp[df_MainPumpsTemp['NmeComp'] == 'MainPumpP1'].sort_values(by='Timestamp')
Df_Pump1.set_index('Timestamp', inplace=True)
resultados_pump1 = treinar_modelos(Df_Pump1, 'MainPumpP1')

# Treinar os modelos para MainPumpP2
Df_Pump2 = df_MainPumpsTemp[df_MainPumpsTemp['NmeComp'] == 'MainPumpP2'].sort_values(by='Timestamp')
Df_Pump2.set_index('Timestamp', inplace=True)
resultados_pump2 = treinar_modelos(Df_Pump2, 'MainPumpP2')

# Concatenar os resultados
SD_ResultGARCH = pd.concat([resultados_pump1, resultados_pump2], axis=1)

# Salvar o DataFrame em um arquivo CSV localmente
SD_ResultGARCH.to_csv('SD_ResultGARCH.csv', index=False)

print("Previs√µes e valores reais salvos em SD_ResultGARCH.csv com sucesso.")

# Agora, carregar o arquivo CSV no GitHub
with open('SD_ResultGARCH.csv', 'rb') as file:
    content = file.read()

# Tentar criar o arquivo no GitHub
try:
    # Verificar se o arquivo j√° existe
    file_content = repo.get_contents(file_path)
    # Se o arquivo existir, atualizar o conte√∫do
    repo.update_file(file_path, "Atualiza SD_ResultGARCH.csv", base64.b64encode(content).decode('utf-8'), file_content.sha)
    print("Arquivo atualizado com sucesso no GitHub.")
except:
    # Se o arquivo n√£o existir, criar um novo arquivo
    repo.create_file(file_path, "Adiciona SD_ResultGARCH.csv", base64.b64encode(content).decode('utf-8'))
    print("Arquivo criado com sucesso no GitHub.")
```
